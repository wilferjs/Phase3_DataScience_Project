**Loading the data into Dataframes**

```python
import pandas as pd
```

```python
#The first data frame has the different characteristics of the respondents
df1= pd.read_csv("/content/training_set_features.csv", index_col="respondent_id", encoding="latin1")
df1.head()
```

```python
#The second dataframe contains a binary classification of whether the respondents got the H1N1 or the Seasonal flu Vaccine
df2= pd.read_csv('/content/training_set_labels.csv',index_col="respondent_id", encoding="latin1")
df2.head()
```

```python
#We concatenate the two dataframes to have one that allows us to view all the columns of each respondents including whether they got H1N1 or Seasonal Flu vaccine
df = pd.merge(df1, df2, on='respondent_id', how='left')
df.head()
```
